```python
feature_names = ["holiday", "workingday", "weather", "temp", "atemp", "humidity", "windspeed_fill", "datetime-year", "datetime-month", "datetime-hour"]
feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)

X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)

X_test.head()
```

```python
y_train = train[["count", "casual", "registered"]]

print(y_train.shape)

y_train.head()
```

## Using RandomForest

### Hyperparameter

```python
from sklearn.ensemble import RandomForestRegressor

# n_estimators
# max_depth
# max_features

model = RandomForestRegressor(random_state = 37)
model
```

```python
## 결과가 랜덤할수 있다, 따라서 고정시키기 위해서는 DecisionTreeRegressor에서 random_state 를 특정숫자(37)로 고정

import numpy as np
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv = 20, scoring = "neg_mean_squared_log_error")

scores =-1 * scores
scores = np.sqrt(scores)
score = scores.mean()

f"Score = {score:.5f}"
```

```python
model.fit(X_train, y_train)
```

```python
import numpy as np
```

```python
label_reg = "registered"

model_reg = RandomForestRegressor(random_state = 37)
model_reg.fit(X_train, train[label_reg])

model_cas = RandomForestRegressor(random_state = 37)
model_cas.fit(X_train, train["casual"])

predictions_reg = model_reg.predict(X_test)
print (label_reg, np.shape(predictions_reg))
predictions_cas = model_cas.predict(X_test)


predictions_count = predictions_reg + predictions_cas
print(np.shape(predictions_count))
predictions_count
```
